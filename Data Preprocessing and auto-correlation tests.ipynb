```python
import pandas as pd
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import pyplot
from pandas import read_csv
# import libraries
```

```python
# open sp500 csv data
df = pd.read_csv('sp500.csv')
# calculate the log differences of daily adjusted close to get daily returns
df['return'] = np.log(df['Adj Close']/df['Adj Close'].shift(1)) 
# drop everything besides returns and date
df.drop(['Open', 'High','Low','Close','Volume','Adj Close'], axis=1,inplace=True)
# drop the first row as it doesn't have any return values
df.drop([0, 0],inplace=True)
```

```python
# divide the data set into 6 different periods
df_1 = df[df['Date']<'1945-01-01']
df_2 = df[(df['Date']>='1945-01-01') & (df['Date']<'1960-01-01')]
df_3 = df[(df['Date']>='1960-01-01') & (df['Date']<'1975-01-01')]
df_4 = df[(df['Date']>='1975-01-01') & (df['Date']<'1990-01-01')]
df_5 = df[(df['Date']>='1990-01-01') & (df['Date']<'2005-01-01')]
df_6 = df[df['Date']>='2005-01-01']
```

```python
# save the data 
List = [df_1,df_2,df_3,df_4,df_5,df_6,df]
for L in List:
    L.drop(['Date'], axis=1,inplace=True)
df.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns.csv',index=False)
df_1.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns_1.csv',index=False)
df_2.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns_2.csv',index=False)
df_3.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns_3.csv',index=False)
df_4.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns_4.csv',index=False)
df_5.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns_5.csv',index=False)
df_6.to_csv(r'C:\Users\Joshua Kim\Desktop\Senior_Comps\returns_6.csv',index=False)
```

```python
# basic statistcal test to see mean and std
List = [df_1,df_2,df_3,df_4,df_5,df_6,df]
for L in List:
    print (L.describe())
```

```python
# autocorrelation test. 1 signals a new time period, and there are 20 autocorrelations (lag 1, lag 2, lag 3, ..., lag 20)
for L in List:
    count=1
    print(count)
    while count<21:
        print(str(L['return'].autocorr(lag=count)))
        count=count+1
# statistically signifant at 5 %
# calculated as ||autocorrelation coefficient|| > 1.96/sqrt(n)
```
